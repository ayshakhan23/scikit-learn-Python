```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('Canada-perCapita-income.csv')
```

```python
df
```

```python
df.rename(columns={'per capita income (US$)': 'PerCapitaIncome'}, inplace=True)
```

```python
df
```

```python
plt.scatter(df.year,df.PerCapitaIncome,color='red',marker='+')
```

```python
new_df = df.drop ('PerCapitaIncome', axis='columns')
```

```python
income = df.PerCapitaIncome
income
```

```python
new_df
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(new_df,income)
```
